```python
import os
import cv2
import torch
from matplotlib import pyplot as plt
from ultralytics import YOLO
import easyocr
```

```python
# Загрузка обученной модели YOLOv8
model_path = "runs/yolo8_experiment4/weights/best.pt"  # Укажите путь к обученной модели
model = YOLO(model_path)

# Инициализация EasyOCR
ocr_reader = easyocr.Reader(['en'], gpu=torch.cuda.is_available())  # Добавьте 'ru', если нужен русский язык

# Папка с тестовыми изображениями
test_images_folder = r"C:\Users\abram\DataspellProjects\FinetuningEasyOCR\dataset\test"  # Укажите путь к тестовой выборке
```

```python
output_folder = "predict_test"  # Папка для сохранения областей
os.makedirs(output_folder, exist_ok=True)
```

```python
# Прогон на тестовой выборке
for image_name in os.listdir(test_images_folder):
    image_path = os.path.join(test_images_folder, image_name)

    if image_name.lower().endswith(('.jpg', '.png', '.jpeg')):
        # Загрузка изображения
        image = cv2.imread(image_path)
        if image is None:
            print(f"Ошибка при загрузке изображения: {image_path}")
            continue

        # Преобразование изображения в RGB
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # YOLO предсказания
        results = model.predict(source=image_rgb, save=False, conf=0.25)  # Установите желаемый порог confidence
        detections = results[0].boxes.cpu().numpy()  # Преобразование в NumPy

        for idx, box in enumerate(detections):
            x_min, y_min, x_max, y_max = map(int, box.xyxy[0])  # Координаты области
            cropped_image = image[y_min:y_max, x_min:x_max]

            # Сохраняем область текста
            cropped_image_path = os.path.join(output_folder, f"{os.path.splitext(image_name)[0]}_box{idx + 1}.jpg")
            cv2.imwrite(cropped_image_path, cropped_image)

            # Распознавание текста
            ocr_results = ocr_reader.readtext(cropped_image)
            for bbox, text, confidence in ocr_results:
                print(f"Распознанный текст: {text}, Уверенность: {confidence:.2f}")

            # Визуализация (опционально)
            plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
            plt.axis('off')
            plt.show()
    else:
        print(f"Пропущен файл: {image_name}")

print(f"Все предсказания сохранены в папку: {output_folder}")
```
